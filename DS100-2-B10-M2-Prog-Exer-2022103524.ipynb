```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

x_values = ['#24','#102','#175','#251','#322','#404','#482','#560','#638','#715']
y_values = [0,44.9629,42.3419,41.461,40.8316,41.3742,41.2468,41.6581,41.5727,41.026] 

md_index = x_values.index('#715')
airbus_index = x_values.index('#24')
x_values[0], x_values[md_index] = x_values[md_index], x_values[0]
x_values[-1], x_values[airbus_index] = x_values[airbus_index], x_values[-1]

fig, ax = plt.subplots()

sorted_data = sorted(zip(x_values, y_values), key=lambda x: x[1], reverse=True)

x_values_sorted, y_values_sorted = zip(*sorted_data)


def update(frame):
    ax.clear()
    data = sorted_data[:frame + 1]  # Take the top 'frame + 1' elements
    x_values_frame, y_values_frame = zip(*data)
    bars = ax.barh(x_values_frame, y_values_frame)  # Use barh for horizontal bars
    plt.title('2017 All Completed Laps', fontweight='bold')
    plt.xlabel('Car Number', fontweight='bold')
    plt.ylabel('Time Difference', fontweight='bold')
    for bar, w in zip(bars, y_values_frame):
        ax.text(w + 5, bar.get_y() + bar.get_height() / 2, str(w), va='center', fontweight='bold')


ani = animation.FuncAnimation(fig, update, frames=len(y_values_sorted), interval=1000, repeat=False)

HTML(ani.to_jshtml())
```
